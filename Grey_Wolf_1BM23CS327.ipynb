```python
import numpy as np

# Example objective function: Sphere function (min at 0)
# Replace this with your actual function f(X)
def objective_function(X):
    return np.sum(X**2)

def Grey_Wolf_Optimizer(N=30, dim=2, max_iter=100, lb=-10, ub=10):

    # Initialize wolf positions
    Wolves = np.random.uniform(lb, ub, (N, dim))
    Fitness = np.array([objective_function(w) for w in Wolves])

    # Identify α, β, δ (best three wolves)
    sorted_idx = np.argsort(Fitness)
    Alpha = Wolves[sorted_idx[0]].copy()
    Beta  = Wolves[sorted_idx[1]].copy()
    Delta = Wolves[sorted_idx[2]].copy()

    Alpha_score = Fitness[sorted_idx[0]]
    Beta_score  = Fitness[sorted_idx[1]]
    Delta_score = Fitness[sorted_idx[2]]

    # Main optimization loop
    for t in range(max_iter):
        a = 2 - (2 * t / max_iter)   # linearly decreases from 2 → 0

        for i in range(N):
            for j in range(dim):

                r1, r2 = np.random.rand(), np.random.rand()
                A1 = 2 * a * r1 - a
                C1 = 2 * r2

                D_alpha = abs(C1 * Alpha[j] - Wolves[i, j])
                X1 = Alpha[j] - A1 * D_alpha

                r1, r2 = np.random.rand(), np.random.rand()
                A2 = 2 * a * r1 - a
                C2 = 2 * r2

                D_beta = abs(C2 * Beta[j] - Wolves[i, j])
                X2 = Beta[j] - A2 * D_beta

                r1, r2 = np.random.rand(), np.random.rand()
                A3 = 2 * a * r1 - a
                C3 = 2 * r2

                D_delta = abs(C3 * Delta[j] - Wolves[i, j])
                X3 = Delta[j] - A3 * D_delta

                # Update wolf's new position
                Wolves[i, j] = (X1 + X2 + X3) / 3

        # Ensure bounds
        Wolves = np.clip(Wolves, lb, ub)

        # Re-evaluate fitness
        Fitness = np.array([objective_function(w) for w in Wolves])

        # Update α, β, δ
        sorted_idx = np.argsort(Fitness)

        if Fitness[sorted_idx[0]] < Alpha_score:
            Alpha = Wolves[sorted_idx[0]].copy()
            Alpha_score = Fitness[sorted_idx[0]]

        if Fitness[sorted_idx[1]] < Beta_score:
            Beta = Wolves[sorted_idx[1]].copy()
            Beta_score = Fitness[sorted_idx[1]]

        if Fitness[sorted_idx[2]] < Delta_score:
            Delta = Wolves[sorted_idx[2]].copy()
            Delta_score = Fitness[sorted_idx[2]]

    return Alpha, Alpha_score


# Run the optimizer
best_pos, best_fit = Grey_Wolf_Optimizer()

print("Best Position (Xα):", best_pos)
print("Best Fitness f(Xα):", best_fit)
```
