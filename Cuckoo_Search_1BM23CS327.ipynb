```python
import numpy as np
import math

# Example objective function: Sphere function (min at 0)
def objective_function(x):
    return np.sum(x**2)

# Correct L√©vy Flight using Mantegna's Algorithm
def levy_flight(lambda_, dim):
    sigma = (math.gamma(1 + lambda_) * math.sin(math.pi * lambda_ / 2) /
            (math.gamma((1 + lambda_) / 2) * lambda_ * 2 ** ((lambda_ - 1) / 2))) ** (1 / lambda_)

    u = np.random.randn(dim) * sigma
    v = np.random.randn(dim)

    step = u / (np.abs(v) ** (1 / lambda_))
    return step

def Cuckoo_Search(n=20, dim=2, max_iter=100, pa=0.25, alpha=0.01, lb=-5, ub=5):

    # Initialize nests
    nests = np.random.uniform(lb, ub, (n, dim))
    fitness = np.array([objective_function(nest) for nest in nests])

    # Identify best nest
    best_index = np.argmin(fitness)
    best_nest = nests[best_index].copy()
    best_fitness = fitness[best_index]

    for t in range(max_iter):
        for i in range(n):

            # L√©vy flight step
            step = levy_flight(1.5, dim)
            new_solution = nests[i] + alpha * step
            new_solution = np.clip(new_solution, lb, ub)

            new_fit = objective_function(new_solution)

            # Choose random nest
            j = np.random.randint(n)

            # Replace if improves
            if new_fit < fitness[j]:
                nests[j] = new_solution
                fitness[j] = new_fit

        # Abandon Pa fraction of nests
        for i in range(n):
            if np.random.rand() < pa:
                nests[i] = np.random.uniform(lb, ub, dim)
                fitness[i] = objective_function(nests[i])

        # Update global best
        best_index = np.argmin(fitness)
        if fitness[best_index] < best_fitness:
            best_fitness = fitness[best_index]
            best_nest = nests[best_index].copy()

    return best_nest, best_fitness


# Run the Algorithm
best_solution, best_value = Cuckoo_Search()

print("‚úÖ Best Solution Found:", best_solution)
print("üèÜ Best Fitness:", best_value)
```
