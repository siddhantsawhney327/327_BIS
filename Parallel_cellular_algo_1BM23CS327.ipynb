```python
import numpy as np

def objective_function(x):
    """Example objective function: Sphere function (min at 0).
       Replace this with your actual f(x)."""
    return np.sum(x**2)

def get_neighbors(index, grid_size):
    """Returns Moore neighborhood indices of a cell index."""
    rows, cols = grid_size
    r, c = index

    neighbors = []
    for dr in [-1, 0, 1]:
        for dc in [-1, 0, 1]:
            if dr == 0 and dc == 0:
                continue
            nr, nc = r + dr, c + dc
            if 0 <= nr < rows and 0 <= nc < cols:
                neighbors.append((nr, nc))
    return neighbors

def Parallel_Cellular_Algorithm(grid_size=(5,5), dim=2, max_iter=100, r=0.5):
    rows, cols = grid_size
    num_cells = rows * cols

    # Initialize population grid
    grid = np.random.uniform(-5, 5, (rows, cols, dim))
    fitness = np.zeros((rows, cols))

    # Evaluate initial fitness
    for i in range(rows):
        for j in range(cols):
            fitness[i, j] = objective_function(grid[i, j])

    # Identify global best
    best_position = grid[fitness.argmin() // cols, fitness.argmin() % cols].copy()
    best_fitness = np.min(fitness)

    # Main Loop
    for _ in range(max_iter):

        new_grid = grid.copy()

        # Parallel cell update (simulated)
        for i in range(rows):
            for j in range(cols):

                neighbors = get_neighbors((i, j), grid_size)

                # Find best neighbor
                best_neighbor = min(neighbors, key=lambda idx: fitness[idx[0], idx[1]])
                br, bc = best_neighbor

                # Update position
                new_position = grid[i, j] + r * (grid[br, bc] - grid[i, j])
                new_fit = objective_function(new_position)

                # Selection Step
                if new_fit < fitness[i, j]:
                    new_grid[i, j] = new_position
                    fitness[i, j] = new_fit

        grid = new_grid  # Update entire grid at once (parallel effect)

        # Update Global Best
        current_best = fitness.min()
        if current_best < best_fitness:
            best_fitness = current_best
            best_pos_idx = np.unravel_index(fitness.argmin(), fitness.shape)
            best_position = grid[best_pos_idx].copy()

    return best_position, best_fitness


# Run Algorithm
best_solution, best_value = Parallel_Cellular_Algorithm()

print("Best Solution Found:", best_solution)
print("Best Fitness:", best_value)
```
